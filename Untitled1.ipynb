```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
format_dict = {'data science':'${0:,.2f}', 'Mes':'{:%m-%Y}', 'machine learning':'{:.2%}'}
#We make sure that the Month column has datetime format
temp1['Mes'] = pd.to_datetime(temp1['Mes'])
#We apply the style to the visualization
temp1.head().style.format(format_dict)
```

```python
format_dict = {'Mes':'{:%m-%Y}'} #Simplified format dictionary with values that do make sense for our data
temp1.head().style.format(format_dict).highlight_max(color='darkgreen').highlight_min(color='#ff0000')
```

```python
temp1.head(10).style.format(format_dict).background_gradient(subset=['data science', 'machine learning'], cmap='BuGn')
```

```python
temp1.head().style.format(format_dict).bar(color='red', subset=['data science', 'deep learning'])
```

```python
temp1.head(10).style.format(format_dict).background_gradient(subset=['data science', 'machine learning'], cmap='BuGn').highlight_max(color='yellow')
```

```python
from pandas_profiling import ProfileReport
prof = ProfileReport(temp1)
prof.to_file(output_file='report.html')
```

```python
import matplotlib.pyplot as plt
plt.plot(temp1['Mes'], temp1['data science'], label='data science') #The parameter label is to indicate the legend. This doesn't mean that it will be shown, we'll have to use another command that I'll explain later.
```

```python
plt.plot(temp1['Mes'], temp1['data science'], label='data science')
plt.plot(temp1['Mes'], temp1['machine learning'], label='machine learning')
plt.plot(temp1['Mes'], temp1['deep learning'], label='deep learning')
```

```python
plt.plot(temp1['Mes'], temp1['data science'], label='data science')
plt.plot(temp1['Mes'], temp1['machine learning'], label='machine learning')
plt.plot(temp1['Mes'], temp1['deep learning'], label='deep learning')
plt.xlabel('Date')
plt.ylabel('Popularity')
plt.title('Popularity of AI terms by date')
plt.grid(True)
plt.legend()
```
